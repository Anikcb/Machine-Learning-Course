## Reading Data

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
x = df.x
y = df.y
```

```python
x.head()
```

## linear regression using gradient descent
[Tutorial of Gradient Descent](https://www.youtube.com/watch?v=Aym1Gx2wVK4&ab_channel=codebasicsHindi)

```python
m = 0;
c = 0;
n = len(y);
learn_rate = 0.0001
for i in range(1000):
    y_pred = (m*x) + c;
    cost = (1/n) * sum([val**2 for val in (y-y_pred)])
    m_der = (-2/n) * sum(x * (y-y_pred))
    c_der = (-2/n) * sum(y - y_pred)
    m = m - learn_rate*m_der
    c = c - learn_rate*c_der
    print("m -> {}   c -> {}   cost -> {}   Iteration -> {}".format(m,c,cost,i))
```

# Multiple Linear Regression
[Tutorial of Multiple Linear Regression](https://youtu.be/QpgJgiB8u7M)

```python
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('agora.csv')
df.head()
```

### Checking Null Values

```python
df.isnull().sum()
```

### Filling Null Values

```python
df.Transport = df.Transport.fillna(df.Transport.mean())
df.isnull().sum()
```

## Encoding - Using Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.Area = le.fit_transform(df['Area'])
df.head()
```

### Spitting Training and Testing Data

```python
from sklearn.model_selection import train_test_split
x = df.drop('Profit',axis=1)
y = df.Profit
xtrain, xtest, ytrain, ytest  = train_test_split(x,y,train_size = .70)
xtrain.head()
```

### Fitting the Model

```python
model = linear_model.LinearRegression()
model.fit(xtrain,ytrain)
```

```python
xtest.head()
```

```python
ytest.head()
```

### Predicting the Data

```python
model.predict([[131876.90,99814.71,362861.36,1]])
```

```python
model.coef_
```

```python
model.intercept_
```

# R - Squared Value

```python
from sklearn.metrics import r2_score
```

```python
y_pred = model.predict(xtest)
```

```python
r2 = r2_score(ytest,y_pred)
r2
```

### Model is better when the r2 value is near to 1
