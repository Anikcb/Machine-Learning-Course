# One Hot Encoding

```python
df = pd.read_csv('agora.csv')
df.head()
```

```python
dummy_var = pd.get_dummies(df['Area'],drop_first=True)
dummy_var.head()
```

```python
df2 = df.drop('Area',axis=1)
```

```python
df2.head()
```

```python
df = df2
```

```python
df.head()
```

```python
df = pd.concat([df,dummy_var],axis=1)
```

```python
df.head()
```

# Ordinal Encoding

```python
df = pd.read_csv('agora.csv')
df.head()
```

```python
city = ['Rangpur','Dhaka', 'Ctg']
```
